```python
import requests
import json

METABASE_URL = 'http://localhost:3001'  # Địa chỉ của Metabase instance của bạn
CARD_ID = 15  # ID của card bạn muốn truy vấn

# Thông tin đăng nhập của bạn
USERNAME = 'metabase@login.com'
PASSWORD = 'metabase123'

# Đăng nhập và lấy token
response = requests.post(
    f"{METABASE_URL}/api/session",
    json={
        "username": USERNAME,
        "password": PASSWORD
    }
)
response.raise_for_status()  # Nếu có lỗi, dừng script
session_token = response.json()['id']

# Sử dụng token để truy vấn thông tin card
headers = {
    "X-Metabase-Session": session_token
}
response = requests.get(
    f"{METABASE_URL}/api/card/{CARD_ID}",
    headers=headers
)
response.raise_for_status()  # Nếu có lỗi, dừng script
card_info = response.json()

# In ra hoặc xử lý thông tin card như bạn muốn
print(json.dumps(card_info, indent=4))

# Nếu bạn muốn, bạn cũng có thể định dạng thông tin này dưới dạng mã Python.
```

```python
import requests

METABASE_URL = 'http://localhost:3001'
USERNAME = 'metabase@login.com'
PASSWORD = 'metabase123'
COLLECTION_NAME = "MVC_collection"

# Đăng nhập và lấy session token
response = requests.post(
    f"{METABASE_URL}/api/session",
    json={
        "username": USERNAME,
        "password": PASSWORD
    }
)
response.raise_for_status()
session_token = response.json()['id']

# Đặt headers cho request kế tiếp
headers = {
    "X-Metabase-Session": session_token
}

# Lấy ID của collection dựa trên tên
response = requests.get(
    f"{METABASE_URL}/api/collection",
    headers=headers
)
response.raise_for_status()
collections = response.json()
collection_id = next((c['id'] for c in collections if c['name'] == COLLECTION_NAME), None)

if not collection_id:
    print(f"Không tìm thấy collection có tên '{COLLECTION_NAME}'")
else:
    # Lấy danh sách các cards thuộc collection
    response = requests.get(
        f"{METABASE_URL}/api/collection/{collection_id}/items",
        headers=headers
    )
    response.raise_for_status()
    cards = response.json()

    # # In ra các card IDs
    # card_ids = [card['id'] for card in cards if card['model'] == 'card']
    # print("Card IDs trong collection:", card_ids)
```
